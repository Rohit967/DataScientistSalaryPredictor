import pandas as pd

import seaborn as sns
import numpy as np

```python
df=pd.read_csv("glassdoor_jobs_cleaned.csv")
```

```python
df.describe()
```

```python
df['Rating'].hist(grid=False)
```

There doesn't seem to be too many ratings on the lower side, and the histogram does seem right skewed.

```python
df.avg_salary.hist(bins=6,grid=False)
```

This plot shows a lot of jobs in the \\$60,000 to \\$100,000 range and an even higher number in the \\$120,000 to \\$140,000 range, but very few jobs in the \\$100,000 to \\$120,000 range, which might indicate a clear distinction in salary between the junior and senior levels.

```python
df.age.hist(bins=15,grid=False)
```

A clearly left-skewed graph, most companies hiring seem to be founded more recently, most of them between 0 and 50 years old. 

```python
df.len_desc.hist(grid=False)
```

Most descriptions for jobs seem to be around 4000 words long, but some descriptions seem to go on for 14,000 words as well!

```python
sns.scatterplot(x=df.len_desc,y=df.age,hue=df.avg_salary)
```

A very cluttered plot, the only thing to focus on are the outliers: 
The really long descriptions are jobs that probably won't pay as much as the others.
The really old companies (200+ years) actually keep their descriptions short, and also pay on a higher scale typically.

```python
df.boxplot(column=['age','avg_salary'])
```

```python
df.boxplot(column='Rating')
```

These boxplots show that age has a lot of outliers on the higher side, rating has a few on the lower side, while average salary has a very uniform distribution.

```python
c1=df[['age','avg_salary','Rating','len_desc','num_comp']].corr()
c1
```

```python
f, ax = plt.subplots(figsize=(8, 5))
mask = np.triu(np.ones_like(c1, dtype=np.bool))
cmap=sns.diverging_palette(220,10,as_cmap=True)
sns.heatmap(c1,vmax=.3,center=0,cmap="YlGnBu",mask=mask,square=True,linewidths=.5,cbar_kws={"shrink": .5},annot=True)
```

Although there are no strong correlations, Description Length, Age and Number of Competitors seem to be slightly correlated.

```python
sns.pairplot(df[['age','avg_salary','Rating','len_desc','num_comp']])
```

```python
df_cat=df[['Location','Headquarters','Size','Type of ownership','Industry','Sector','Revenue','Company Name',
           'State','python_yn','r_yn','aws_yn','spark_yn', 'hadoop_yn', 'docker_yn',
       'sql_yn', 'linux_yn', 'flask_yn', 'django_yn', 'tensorflow_yn',
       'keras_yn', 'pytorch_yn', 'algo_yn', 'stats_yn', 'tableau_yn',
       'job_type', 'seniority']]
```

```python
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

```python
for i in df[['Location','Headquarters','Industry','Company Name']].columns:
    cat_num = df_cat[i].value_counts()[:20]
    print("graph for %s: The Top %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

```python
pd.pivot_table(df,index='job_type',values='avg_salary',aggfunc=[np.min,np.max,np.mean,np.std])
```

Data Scientists have an average salary of \\$116,000 while Data Engineers have an average salary of \\$101,000. Analysts, have an average salary of \\$99,000. Surprisingly, Directors and Managers are being paid slightly less, although this could be a fault with the estimate that Glassdoor has provided.

```python
pd.pivot_table(df,index=['job_type','seniority'],values='avg_salary')
```

In most cases, it seems like a senior level pays a little bit lesser. Again, this could be because this data is taken during the COVID-19 pandemic, and lesser senior level and more junior level people are being hired.

```python
pd.pivot_table(df,index='State',values='avg_salary',margins=True,aggfunc=[np.mean,'count'])
```
